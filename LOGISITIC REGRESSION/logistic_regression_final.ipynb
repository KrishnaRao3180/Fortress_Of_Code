```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2,
                           n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=14)

print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
```

```python
lor = LogisticRegression(penalty=None, solver='sag')
lor.fit(X, y)
```

```python
print(lor.intercept_)
print(lor.coef_)
print(lor.intercept_.shape)
print(lor.coef_.shape)
```

```python
m_lor = -(lor.coef_[0][0]/lor.coef_[0][1])
c_lor = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m_lor*x_input + c_lor
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.plot(x_input, y_input, color='black', label='sklearn_model')
plt.ylim(-3.5, 2.5)
plt.legend()
plt.show()
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def GD(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.01
    epoch = 25000

    for i in range(epoch):
        y_hat = sigmoid(np.dot(X, weights))
        weights = weights + lr*(np.dot((y-y_hat), X)/X.shape[0])

    return weights[1:], weights[0]
```

```python
coef_my_model, intercept_my_model = GD(X, y)
print(coef_my_model.shape)
print(intercept_my_model.shape) 
m_mymodel = -(coef_my_model[0]/coef_my_model[1])
c_mymodel = -(intercept_my_model/coef_my_model[1])
```

```python
y_input_mymodel = m_mymodel*x_input + c_mymodel
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.plot(x_input, y_input, color='black', label='sklearn_model')
plt.plot(x_input, y_input_mymodel, color='purple', label='my_model')
plt.ylim(-3.5, 2.5)
plt.legend()
plt.show()
```
