```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=50, random_state=13)
```

```python
plt.scatter(X, y)
```

```python
reg = LinearRegression()
reg.fit(X, y)
print(reg.coef_)
print(reg.intercept_)
plt.scatter(X, y, color='blue')
plt.plot(X, reg.predict(X), color='red')
plt.show()
```

```python
# Now nowing the slope fixed as 71.80 we will us GD to find the optimal value of b  
# Lets start with the starting value of b as b = 0

y_pred = (71.80 * X + 0)
print(y_pred)
y_pred = y_pred.reshape(4)
print(y_pred)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color="#30835e", label='b = 0')
plt.legend()
plt.show()
```

```python
m = 71.70
b = 100                   # the initial value of b to start with
# print(y)
# print(X.reshape(4))
# print(-2 * sum(y - m*X.reshape(X.size)))
loss_slope = -2 * sum(y - m*X.ravel() - b)  
```

```python
lr = 0.1            # LEARNING RATE
step_size = lr*loss_slope
b -= step_size
```

```python
y_pred = 71.80 * X + 0
y_pred1 = 71.80 * X + b

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color="#30835e", label='b = 0')
plt.plot(X, y_pred1, color='#A3E4D7', label=f"{b}")
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2*sum(y - 71.80*X.ravel() - b)
b -= loss_slope*lr
y_pred2 = 71.80 * X + b

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color="#30835e", label='b = 0')
plt.plot(X, y_pred1, color="#A3E4C2", label=f"13.08952957")
plt.plot(X, y_pred2, label=f'{b}')
plt.legend()
plt.show()
```

```python
for i in range(10):
    loss_slope = -2 * sum(y - 71.80*X.ravel() - b)
    b -= loss_slope*lr

print(b)
print(reg.intercept_)

y_pred_final = 71.80*X + b


plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color="#30835e", label='b = 0')
plt.plot(X, y_pred1, color="#A3E4C2", label=f"13.08952957")
plt.plot(X, y_pred2, label=f'15.69846179')
plt.plot(X, y_pred_final, color='black', label=f'{b}')
plt.legend()
plt.show()
```

```python
b = -50
lr = 0.01
epochs = 50

for i in range(epochs):
    loss_slope = -2 * sum(y - 71.80*X.ravel() - b)
    b -= loss_slope*lr

    y_pred = 71.80*X + b
    plt.plot(X, y_pred)

plt.plot(X, reg.predict(X), color='red')
plt.scatter(X, y, color='blue')
```
