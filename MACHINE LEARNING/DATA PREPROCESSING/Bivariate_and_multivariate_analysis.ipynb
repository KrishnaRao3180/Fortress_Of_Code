```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
titanic = pd.read_csv("K:\\PYTHON\\MACHINE LEARNING\\Fortress_Of_Code\\DATASET\\Titanic-Dataset.csv")
iris = sns.load_dataset("iris")
flight = sns.load_dataset("flights")
tips = sns.load_dataset("tips")
```

```python
print(titanic.shape)
print(iris.shape)
print(flight.shape)
print(tips.shape)
```

```python
# Using Scatterplot to see the relationship between numerical and numerical columns

sns.scatterplot(x = tips[tips['size'] > 2]['total_bill'], y = tips[tips['size'] > 2]['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'])
```

```python
#  Using barplot to see the relationship between Categorical and Numerical Columns

plt.figure()
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'], hue=titanic['Sex'])

plt.figure()
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'], hue=titanic['Embarked'])

plt.show()
```

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Fare'], hue=titanic['Pclass'])
```

```python
sns.boxplot(x=titanic[titanic['Sex']=='male']['Sex'], y=titanic['Fare'], hue=titanic['Pclass'])
```

```python
# Using distplot() for seeing the probability distribution of something 
# Especially in the titanic dataset we would be intersted in the probability of survival

plt.figure()
sns.distplot(titanic['Age'])

plt.figure()
sns.distplot(titanic[titanic["Survived"] == 1]['Age'], hist=False)
sns.distplot(titanic[titanic["Survived"] == 0]['Age'], hist=False)
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```
