```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2,
                           n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)

print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
```

```python
def step(z):
    return 1 if z > 0 else 0
```

```python
def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for j in range(1000):
        j = np.random.randint(0, X.shape[0])
        y_hat = step(np.dot(X[j], weights))
        weights = weights + (y[j] - y_hat)*X[j]
    
    return weights[0], weights[1:]
```

```python
intercept_, coeff_ = perceptron(X, y)
```

```python
print(coeff_)
print(intercept_)
```

```python
m = -(coeff_[0]/coeff_[1])
c = -(intercept_/coeff_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + c
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.plot(x_input, y_input)
plt.ylim(-4,3)
```

```python
lor = LogisticRegression()
lor.fit(X, y)
```

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
x_input1 = np.linspace(-3, 3, 100)
y_input1 = m1*x_input + b
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.plot(x_input1, y_input1, label='Sklearn_line', color='black')
plt.plot(x_input, y_input, label='Custom_model')
plt.ylim(-4,3)
plt.legend()
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def perceptron_imporoved(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for j in range(1000):
        j = np.random.randint(0, X.shape[0])
        y_hat = sigmoid(np.dot(X[j], weights))
        weights = weights + (y[j] - y_hat)*X[j]
    
    return weights[0], weights[1:]
```

```python
intercept_imp, coeff_improved = perceptron_imporoved(X, y)
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
m2 = -(coeff_improved[0]/coeff_improved[1])
b2 = -(intercept_imp/coeff_improved[1])
y_input1 = m1*x_input + b
y_input2 = m2*x_input + b2
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.plot(x_input1, y_input1, label='Sklearn_line', color='black')
plt.plot(x_input, y_input, label='Custom_model')
plt.plot(x_input, y_input2, label='Custom_model_improved')
plt.ylim(-4,3)
plt.legend()
```
