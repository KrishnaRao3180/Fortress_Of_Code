```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("K:\\PYTHON\\MACHINE LEARNING\\LINEAR REGRESSION\\DATA\\placement.csv")
print(df.head())
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.show()
```

```python
# Segregating the input and output values

X = df.iloc[:, 0:1]
Y = df.iloc[:, -1]

print(X.head())
print(Y.head())

print(X.iloc[0:2].values.reshape(1, 2))
print(X.iloc[0:2].values)
print(X.iloc[0:2].values.shape)
print(Y.iloc[2])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
print(X_test.head())
print()
print(Y_test.head())
```

```python
lr.predict(X_test.iloc[0].values.reshape(1, 1))
```

```python
lr.predict(X_test.iloc[1].values.reshape(1,1))
```

```python
# print(lr.predict(X_train))
plt.plot(X_train, lr.predict(X_train),color ='red')
plt.plot(np.arange(1, 8), lr.coef_*np.arange(1, 8) + lr.intercept_ + 2, color = 'orange')
plt.scatter(X, Y, color = 'blue')
plt.scatter(X_test, Y_test, color = 'green')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.show()
```

```python
print(lr.coef_)     # This is the slope of the line which is also called weightage telling how much the 
# quantity depends on each other higher slope means higher value change for a small change in the value

print(lr.intercept_) # This is the biasis of the quantities which sets some offset for the model

plt.plot(np.arange(1, 100), lr.coef_*np.arange(1, 100) + lr.intercept_)
plt.show()
```
