```python
def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    epoch = 1000                        # BTW 2 epochs is all you need for this dataset and values

    for i in range(epoch):
        for j in range(X.shape[0]):
            z = np.dot(weights, X[j])
            for k in range(weights.shape[0]):
                if y[j]*z < 0:
                    derivative = - y[j]*X[j][k]
                    weights[k] -= lr*derivative


    return weights[1:], weights[0]
```

```python
weights, bias = perceptron(X, y)
```

```python
m = -(weights[0]/weights[1])
c = -(bias/weights[1])

print(m, c)
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + c
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.plot(x_input, y_input, color='black', label='sklearn_model')
plt.ylim(-3.5, 2.5)
plt.legend()
plt.show()
```
